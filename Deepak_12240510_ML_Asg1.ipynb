**Q1: Remove Outliers and Compare Dataset Sizes**

```python
import pandas as pd

df = pd.read_csv('/content/zomato.csv', encoding='ISO-8859-1')
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    df_no_outliers = df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]
    return df_no_outliers

df_cleaned = remove_outliers_iqr(df, 'Aggregate rating')
original_size = df.shape
cleaned_size = df_cleaned.shape

print(f"Original dataset size: {original_size}")
print(f"Cleaned dataset size: {cleaned_size}")
```

**4. Explore Whether We Can Merge "zomato.csv" and "country_code.xlsx"**

```python
import pandas as pd

# Load the datasets
zomato = pd.read_csv('/content/zomato.csv', encoding='ISO-8859-1')
country_df = pd.read_excel('/content/Country-Code.xlsx')

# Merge the datasets
merged_df = pd.merge(zomato, country_df, on='Country Code', how='inner')

# Save the merged dataset to CSV
merged_df.to_csv('/content/merged_zomato.csv', index=False)
```
